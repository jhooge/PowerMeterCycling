```python
import os
import glob
import datetime
from dateutil import tz, relativedelta

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from fitparse import FitFile

sns.set(style="whitegrid")
```

```python
class MissingSensorError(Exception):
    def __init__(self, value):
        self.value = "Missing %s sensor data."%value
    def __str__(self):
        return repr(self.value)

class FitnessProfile:
    def __init__(self, ftp, weight, age, 
                 height=179, gender="male", name="Jens Hooge",
                 prefered_cadence=80):
        self.name = name
        self.gender = gender
        self.height = height
        self.ftp = ftp
        self.prefered_cadence = prefered_cadence
        self.weight = float(weight)
        self.age = age
        self.power_zones = self.pwr_zones()
        self.heart_rate_zones = self.hr_zones()
        
    def pwr_zones(self):
        '''Coggan power zones 
          (see: https://www.trainingpeaks.com/blog/power-training-levels/)'''
        zones = {
                 'Active Recovery':self.ftp*0.56,   ## upper limit
                 'Endurance':self.ftp*0.76,         ## upper limit
                 'Tempo':self.ftp*0.91,             ## upper limit
                 'Lactate Threshold':self.ftp*1.06, ## upper limit
                 'VO2max':self.ftp*1.21,            ## upper limit
                 'Anaerobic Capacity':2000
                }
        return(zones)
    
    def hr_zones(self):
        '''How are these zones calculated in Wahoo ELEMNT'''
        zones = {"peak":188,
                 "hard":(168, 187),
                 "cardio":(148, 167),
                 "fat burn":(129, 147),
                 "easy":(0, 128)}
        return(zones)
    
    def power_to_weight_ratio():
        pass
    
class Workout:
    
    def __init__(self, data):
        
        self.data = data
        self._preprocess()
        self.has_power = 'power' in self.data.columns
        self.has_heart_rate = 'heart_rate' in self.data.columns
        self.has_cadence = 'cadence' in self.data.columns
        self.has_speed = 'speed' in self.data.columns
        self.has_temp = 'temperature' in self.data.columns
        self.has_altitude = 'enhanced_altitude' in self.data.columns
        self.has_position_long = 'position_long' in self.data.columns
        self.has_position_lat = 'position_lat' in self.data.columns
        
        self.position_long = None
        self.position_lat = None
        self.altitude = None
        
        ## performance summary
        self.avg_power = None
        self.max_power = None
        self.min_power = None
        self.norm_power = None
        
        self.avg_hr = None
        self.max_hr = None
        self.min_hr = None
        
        self.min_temp = None
        self.max_temp = None
        self.avg_temp = None

        ## workout summary
        self.timestamp = self.data.timestamp.min()
        self.distance = self.data.distance.max()
        self.duration = self._workout_duration()
        
        if self.has_position_long:
            self.position_long = self.data.position_long
            
        if self.has_position_lat:
            self.position_lat = self.data.position_lat
            
        if self.has_altitude:
            self.altitude = self.data.enhanced_altitude
            self.ascent = self.data.enhanced_altitude.max()
        
        if self.has_temp:
            self.min_temp = self.data.temperature.min()
            self.max_temp = self.data.temperature.max()
            self.avg_temp = self.data.temperature.mean()
        
        if self.has_power:
            self.avg_power = self.data.power.mean()
            self.max_power = self.data.power.max()
            self.min_power = self.data.power.min()
            self.norm_power = self._normalize_power(self.data.power)
        
        if self.has_heart_rate:
            self.avg_hr = self.data.heart_rate.mean()
            self.max_hr = self.data.heart_rate.max()
            self.min_hr = self.data.heart_rate.min()
            
        if self.has_cadence:
            self.avg_cad = self.data.cadence.mean()
            self.max_cad = self.data.cadence.max()
            self.min_cad = self.data.cadence.min()
            
        if self.has_speed:
            self.avg_speed = self.data.speed.mean()
            self.max_speed = self.data.speed.max()
            self.min_speed = self.data.speed.min()
            
        
    def __str__(self):
        string = "Workout on %s\n" %str(self.timestamp)
        string += "---------------\n"
        string += "Duration: %s\n" %str(datetime.timedelta(seconds=self.duration))
        string += "Distance: %.2f km\n" %self.distance
        
        if self.has_altitude:
            string += "Ascent: %.2f m\n" %self.ascent
        
        string += "\n"
        
        if any([self.has_power, self.has_heart_rate, self.has_cadence, 
                self.has_speed, self.has_temp]):
            string += "             Min.\tMax.\tAvg.\n"
            if self.has_power:
                string += "Power        %.2f\t%.2f\t%.2f watts\n" %(self.min_power, self.max_power, self.avg_power)
            if self.has_heart_rate:
                string += "Heart Rate     %i\t   %i\t   %i bpm\n" %(self.min_hr, self.max_hr, self.avg_hr)
            if self.has_cadence:
                string += "Cadence      %.2f\t%.2f\t %.2f rpm\n" %(self.min_cad, self.max_cad, self.avg_cad)
            if self.has_speed:
                string += "Speed         %.1f\t  %.1f\t  %.1f kmh\n" %(self.min_speed, self.max_speed, self.avg_speed)
            if self.has_temp:
               string += "Temperature  %.1f\t   %.1f\t  %.1f C\n" %(self.min_temp, self.max_temp, self.avg_temp)
        
        return(string)

    def _workout_duration(self):
        duration = self.data.timestamp.max() - self.data.timestamp.min()
        duration = duration.total_seconds()
        return(duration)
    
    ## Preprocessing
    def _convert_speed(self, x):
        '''Converts speed from m/s to km/h'''
        y = x*3.6
        return(y)

    def _convert_distance(self, x):
        '''Converts distance to km'''
        y = x/1000.
        return(y)

    def _convert_longlat(self, x):
        '''Converts semicircles to degrees'''
        y = x*(180/2^31)
        return(y)

    def _preprocess(self):
        self.data.distance = self._convert_distance(self.data.distance)
        
        if 'enhanced_altitude' in self.data.columns:
            self.data.altitude = self.data.enhanced_altitude
        
        if 'speed' in self.data.columns:
            self.data.speed = self._convert_speed(self.data.enhanced_speed)
        
        if 'position_lat' in self.data.columns:
            self.data.position_lat = self._convert_longlat(self.data.position_lat)
        
        if 'position_long' in self.data.columns:
            self.data.position_long = self._convert_longlat(self.data.position_long)
        
        ## sometimes the first few sensor measurements are NaN, so
        ## we will replace them by 0
        
        for col in ['power', 'speed', 'cadence']:
            if col in self.data.columns:
                self.data.loc[:20, col] = self.data.loc[:20, col].fillna(0)
        
        return(None)

    def _normalize_power(self, power, win_size=30, zeroes=True):
        '''
        Normalize power (see: Coggan et al)
        This seems to underestimate the NP of my Wahoo. TODO: Why?
        '''
        norm_power = smoothing(power, win_size)
        norm_power = np.power(norm_power, 4)
        norm_power = norm_power.mean()
        norm_power = np.power(norm_power, 1./4)
        return(norm_power)
    
    def intensity_factor(self, ftp):
        ifac = None
        if self.has_power:
            ifac = self.norm_power/ftp
        else:
            raise MissingSensorError("power")
            
        return(ifac)
    
    def variability_index(self):
        vi = None
        if self.has_power:
            vi = self.norm_power/self.avg_power
        else:
            raise MissingSensorError("power")
            
        return(vi)

    def training_stress(self, ftp):
        tss = None
        if self.has_power:
            tss = ((self.duration*self.norm_power*self.intensity_factor(ftp))/(ftp*3600))*100
            
        return(tss)
    
    def training_stress_level(self, ftp):
        '''Returns a typical exertion levels for an intensity factor'''
        tss = self.training_stress(ftp)
        
        stress_level = None
        if (tss < 150): 
            stress_level = "Low"
        if((tss >= 150) & (self.intensity_factor(ftp) < 300)):
            stress_level = "Moderate"
        if((tss >= 300) & (self.intensity_factor(ftp) < 450)):
            stress_level = "High"
        if((tss >= 450)):
            stress_level = "Very high"
        return(stress_level)
    
    def excertion_level(self, ftp):
        '''Returns a typical exertion levels for an intensity factor'''
        ifac = self.intensity_factor(ftp)
        
        exertion = None
        if ifac != None:
            if (ifac < 0.75): 
                exertion = "Level 1, Active Recovery"
            if((ifac >= 0.75) & (ifac < 0.85)):
                exertion = "Level 2, Endurance"
            if((exertion >= 0.85) & (ifac < 0.95)):
                exertion = "Level 3, Tempo"
            if((exertion >= 0.95) & (ifac < 1.05)):
                exertion = "Level 4, Lactate Threshold"
            if((exertion >= 1.05)):
                exertion = "Level 5, and higher"
                
        return(exertion)
    
    def power_profile(self, intervals=[1,5,10,30,60,300,600,1800,3600], weight=None):
        '''Computes power profile over time intervals
        TODO: Find out how this is calculated
        TODO: x-axis ticks should be custom spaced (see Wahoo app)
        '''
        intervals = np.array(intervals)
        powers = np.full(len(intervals), np.nan)
        profile_dict = {"interval":intervals, "power":powers}
        profile = pd.DataFrame.from_dict(profile_dict)
        
        if self.has_power:
            powers = map(lambda x: np.max(self.data.power.rolling(x).mean()), intervals)
            profile_dict = {"interval":intervals, "power":powers}
            if weight != None:
                profile_dict = {"interval":intervals, "power":np.array(powers)/float(weight)}
            profile = pd.DataFrame.from_dict(profile_dict)
            
        return(profile)

def smoothing(x, win_size=30):
        y = x.rolling(win_size).mean()
        return(y)    

def fit_to_df(fitfile_path):
    fitfile = FitFile(fitfile_path)
    records = [record for record in fitfile.get_messages('record')]
    records = map(lambda x: x.get_values(), records)
    data = pd.DataFrame(records[1:])
    return(data)
    
## Metrics

def time_in_power_zones(workout, power_zones):
    
    power_zones_df = None

    if workout.has_power:
        in_active_recovery = np.sum(workout.data.power <= power_zones['Active Recovery'])
        in_endurance = np.sum((workout.data.power > power_zones['Active Recovery']) & 
                              (workout.data.power <= power_zones['Endurance']))
        in_tempo = np.sum((workout.data.power > power_zones['Endurance']) & 
                          (workout.data.power <= power_zones['Tempo']))
        in_lac_thresh = np.sum((workout.data.power > power_zones['Tempo']) & 
                               (workout.data.power <= power_zones['Lactate Threshold']))
        in_vo2max = np.sum((workout.data.power > power_zones['Lactate Threshold']) & 
                           (workout.data.power <= power_zones['VO2max']))
        in_anaerob_cap = np.sum((workout.data.power > power_zones['VO2max']))

        power_zones_df = pd.DataFrame.from_dict({"Zone":["Active\nRecovery", "Endurance", "Tempo", 
                                                      "Lactate\nThreshold", "VO2max", "Anaerobic\nCapacity"],
                                                      "Percentage":np.array([in_active_recovery, in_endurance, 
                                                               in_tempo, in_lac_thresh,
                                                               in_vo2max, in_anaerob_cap])/float(len(workout.data.power))*100})
    else:
        raise MissingSensorError("power")
    
    
    return(power_zones_df)

def avg_eff_pedal_force(P, C):
    '''
    The average effective pedal force in newton (AEPF) estimates 
    the velocity of muscle contraction.
    
    AEPF is average effective pedal force (in newtons, or N); 
    P is power (in watts); 
    C is cadence (in rpm); CL is crank length (in meters); 
    and the constants 60, 2, and π serve to convert cadence to 
    angular velocity (in radians/seconds)
    '''
    
    aepf = np.array(P/circ_pedal_velocity(C))
    aepf[np.isnan(aepf)] = 0.0 ## if pedals not turning, power should be zero
    aepf[np.isinf(aepf)] = 0.0 ## if pedals not turning, power should be zero
    return(aepf)

def circ_pedal_velocity(C, CL=0.17):
    '''
    CPV is circumferential pedal velocity (in meters/second); 
    C is cadence (in rpm); 
    CL is crank length (in meters); 
    and the constants 2, π, and 60 serve to convert the data to 
    the proper units.
    '''
    cpv = (C*CL*2*np.pi)/60.0
    return(cpv)

def chronic_training_load(tss_history):
    '''
    Chronic Training Load (CTL) or fitness. Taking into consideration both volume 
    and intensity, CTL provides a measure of how much an athlete has 
    been training historically, or chronically. It is calculated as an 
    exponentially weighted moving average of daily TSS (or TRIMP) values, 
    with the default time constant set to 42 days.“
    
    Auszug aus: Hunter Allen and Andrew Coggan, PhD. 
    „Training and Racing with a Power Meter: 2nd Edition.“ iBooks. 
    '''
    CTL = tss_history['TSS'].ewm(span=42).mean()
    return(CTL)

def acute_training_load(tss_history):
    '''
    Acute Training Load (ATL) or fatigue. Again, taking into consideration both volume 
    and intensity, ATL provides a measure of how much an athlete has been 
    training recently, or acutely. It is calculated as an exponentially 
    weighted moving average of daily TSS values, with the default time 
    constant set to 7 days.“'

    Auszug aus: Hunter Allen and Andrew Coggan, PhD. 
    „Training and Racing with a Power Meter: 2nd Edition.“ iBooks. 
    '''
    ATL = tss_history['TSS'].ewm(span=7).mean()
    return(ATL)

def training_stress_balance(tss_history):
    '''
    Training Stress Balance (TSB). As the name suggests, Training Stress Balance 
    (a term coined by cycling coach Dave Harris) is the difference between chronic 
    and acute training loads: TSB = CTL – ATL. TSB provides a measure of how much 
    an athlete has been training recently, or acutely, compared with how much he 
    or she has trained historically, or chronically. “

    Auszug aus: Hunter Allen and Andrew Coggan, PhD. 
    „Training and Racing with a Power Meter: 2nd Edition.“ iBooks. 
    '''
    TSB = chronic_training_load(tss_history)-acute_training_load(tss_history)
    return(TSB)

def calculate_form(tss_history):
    form = chronic_training_load(tss_history) + training_stress_balance(tss_history)
    return(form)
```

## Overall Training History

```python
def create_training_history(workouts):
    train_hist = pd.DataFrame.from_dict({"TSS":[w.training_stress(ftp) for w in workouts],
                                         "Duration":[w.duration for w in workouts],
                                         "Distance":[w.distance for w in workouts]})
    train_hist.index = [w.timestamp for w in workouts]
#    train_hist = train_hist.groupby(train_hist.index.date).sum()
    train_hist = train_hist.groupby(train_hist.index.date).sum()    

#    origin_date = datetime.datetime(2018, 11, 20)
    first_workout_date = workouts[0].timestamp.date()
    current_date = datetime.datetime.today().date()
    index = pd.date_range(start=first_workout_date, end=current_date, closed=None)

    train_hist = train_hist.reindex(index)
    train_hist["Date"] = train_hist.index
    train_hist = train_hist.fillna(0)
    
    train_hist['CTL'] = chronic_training_load(train_hist)
    train_hist['ATL'] = acute_training_load(train_hist)
    train_hist['TSB'] = training_stress_balance(train_hist)
    train_hist['dCTL'] = train_hist['TSS'] - train_hist['CTL'] # TSS above, below CTL
    train_hist['Form'] = calculate_form(train_hist)
    return(train_hist)

'''
To increase Fitness (CTL) your TSS should be higher than CTL more often than not during training weeks.
In recovery weeks TSS should be lower.

Rule of thumb:
Load weeks: CTL + 30 TSS/day
Revovery weeks: CTL - 20 TSS/day
Theoretical Ideal: 10 CTL/month ramp rate
'''

def training_summary(training_history):
    string = ""
    string += "Training Summary\n"
    string += "Total Distance: %.2f\n"% training_history.Distance.sum()
    string += "Total Duration: %s\n"%str(datetime.timedelta(seconds=training_history.Duration.sum()))
    string += "Total TSS: %.2f\n\n"% training_history.TSS.sum()
    string += str(training_history.describe())
    print(string)
    return(None)
    
def plot_training_history(training_history):
    
    (fig, ax) = plt.subplots(figsize=(20,10), nrows=1)
    
    fig = sns.lineplot(x="Date", y="Form", label='Form', color='red', data=training_history, ax=ax)
    fig = sns.lineplot(x="Date", y="TSS", label='Training Stress (TSS)', 
                       color='orange', linestyle='--', alpha=0.7, 
                       data=training_history, ax=ax)
    ax.lines[1].set_linestyle("--")
    fig = sns.lineplot(x="Date", y="CTL", label='Fitness (CTL)', 
                       color='green', linestyle='--',  alpha=0.7, 
                       data=training_history, ax=ax)
    ax.lines[2].set_linestyle("--")
    fig = sns.lineplot(x="Date", y="ATL", label='Fatigue (ATL)', 
                       color='gray', linestyle='--',  alpha=0.7, 
                       data=training_history, ax=ax)
    ax.lines[3].set_linestyle("--")
    ax.text(training_history.Date.min() , training_history.TSS.max()-10 ,
            'Total TSS: %.2f'%training_history.TSS.sum(), fontsize=16)
    ax.text(training_history.Date.min() , training_history.TSS.max()-20 ,
            'Avg. TSS: %.2f'%training_history.TSS.mean(), fontsize=16)
    ax.set(xlabel='', ylabel='')

#    plt.xticks(rotation=90)
    return(fig)

def plot_freshness(training_history):
    (fig, ax) = plt.subplots(figsize=(20,3), nrows=1)
    fig = sns.barplot(x="Date", y="TSB", label='Form (TSB)', color='lightgray', data=training_history, ax=ax)
    ax.axhline(0, color='lightgray') ## Normalized power
    ax.set(xlabel='', ylabel='', xticks=[], title='Freshness Balance')
    return(fig)

def plot_training_stimulus(training_history):
    (fig, ax) = plt.subplots(figsize=(20,3), nrows=1)
    fig = sns.barplot(x="Date", y="dCTL", label='Diff. CTL', data=training_history, color='lightgray', ax=ax)
    ax.axhline(0, color='lightgray') ## Normalized power
    ax.set(xlabel='', ylabel='', xticks=[], title='Training Stimulus')
    return(fig)
    
```

```python
import matplotlib

def plot_time_in_power_zones(df):
    (fig, ax) = plt.subplots(figsize=(10,5), ncols=1, nrows=1)
    fig = sns.barplot(x="Zone", y="Percentage", data=df, 
                      palette='coolwarm',
                      ax=ax)
    ax.set(xlabel='', ylabel='Time spent in Zone [%]')
    return(fig)

def plot_power_curve_history(workouts, weight=None, 
                             intervals=[1,5,10,30,60,300,600,1800,3600]):

    (fig, ax) = plt.subplots(figsize=(10,5), ncols=1, nrows=1)
    
    power_profiles = np.array(map(lambda w: w.power_profile(intervals).power, workouts))
    overall_max_power = np.nanmax(power_profiles, axis=0)
    overall_mean_power = np.nanmean(power_profiles, axis=0)
    
    if weight != None:
        overall_max_power /= weight
        overall_mean_power /= weight
        power_profiles /= weight
    
    fig = sns.lineplot(x=intervals, y=overall_max_power, 
                      color='black', label="Overall max. power",
                      ax=ax)
    fig = sns.lineplot(x=intervals, y=overall_mean_power, 
                      color='gray', label="Overall avg. power",
                      ax=ax)
    fig = sns.lineplot(x=intervals, y=power_profiles[-1], 
                       color='orange', label="Last workout power",
                       ax=ax)
    ax.lines[0].set_linestyle("--")
    ax.lines[1].set_linestyle("--")
    ax.set(xlabel='Interval [s]', ylabel='Power [Watt]', xscale="log")
    if weight != None:
        ax.set(xlabel='Interval [s]', ylabel='Power [Watt/kg]', xscale="log")
        
    ax.set_xticks(intervals)
    ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
    return(fig)

def plot_power_curve(workout, weight=None, 
                     intervals=[1,5,10,30,60,300,600,1800,3600]):
    plt.ioff()
    
    (fig, ax) = plt.subplots(figsize=(10,5), ncols=1, nrows=1)
    power_profile = workout.power_profile(intervals).power
    
    if weight != None:
        power_profile /= weight
    
    ax = sns.lineplot(x=intervals, y=power_profile, 
                 palette='coolwarm', label="Workout power",
                 ax=ax)
    
    ax.set(xlabel='Interval [s]', ylabel='Power [Watt]', xscale="log")
    if weight != None:
        ax.set(xlabel='Interval [s]', ylabel='Power [Watt/kg]', xscale="log")
        
    ax.set_xticks(intervals)
    ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
    
    return(fig)
    

for i in range(1, len(workouts)):
    plot_power_curve_history(workouts[:i])


#if not workout.has_power:
#    print("WARNING: Not Data to Display!")
#    print("TODO: Show time spent in heart rate zones")
#else:
#    power_zones_df = time_in_power_zones(workout, power_zones)
#    power_df = workout.power_profile()
#    power_weight_df = workout.power_profile(fitness_profile.weight)

#    plot_time_in_power_zones(power_zones_df)
#    plot_power_curve(power_df)
#    plot_power_weight_curve(power_weight_df)
    

## overall max power profile
#print(np.nanmax(np.array(map(lambda w: w.power_profile().power, workouts)), axis=0))

```

```python
## Force vs. Velocity
## Quadrant I: High Force and high velocity (upper right)
## - Sprinting Quadrant
## - Efforts above personal FTP (attacks, bridge attempts)
## - Typical profile of track races

## Quadrant II: High Force and low velocity (upper left)
## - Climbing Quadrant
## - Acceleration from low speed
## - Typical in cyclocross/MTB races

## QI/Q2: Fast twitch recruitment

## Quadrant III: Low Force and low velocity (lower left)
## - Recovery Quadrant
## - Typical by soft-pedalling in large bunch

## Quadrant IV: Low Force and high velocity (lower right)
## - Crit Quadrant
## - Typical for fixed gear efforts

def plot_aepf_cvp_dist(data, ftp, pref_cad):
    
    cpv = circ_pedal_velocity(data.cadence)
    aepf = avg_eff_pedal_force(data.power, data.cadence)
    pedal_behaviour = pd.DataFrame({'aepf':aepf, 'cpv':cpv})
    
    power_curve_at_ftp = avg_eff_pedal_force(ftp, data.cadence)
    cpv_at_pref_cadence = circ_pedal_velocity(prefered_cadence)
    aepf_at_ftp = avg_eff_pedal_force(ftp, prefered_cadence)
    
    time_in_quadrant_1 = np.sum((pedal_behaviour.aepf > aepf_at_ftp) & 
                                (pedal_behaviour.cpv > cpv_at_pref_cadence))
    time_in_quadrant_1  = (time_in_quadrant_1/float(pedal_behaviour.shape[0]))*100
    
    time_in_quadrant_2 = np.sum((pedal_behaviour.aepf > aepf_at_ftp) & 
                                (pedal_behaviour.cpv <= cpv_at_pref_cadence))
    time_in_quadrant_2  = (time_in_quadrant_2/float(pedal_behaviour.shape[0]))*100
    
    time_in_quadrant_3 = np.sum((pedal_behaviour.aepf <= aepf_at_ftp) & 
                                (pedal_behaviour.cpv <= cpv_at_pref_cadence))
    time_in_quadrant_3  = (time_in_quadrant_3/float(pedal_behaviour.shape[0]))*100
    
    time_in_quadrant_4 = np.sum((pedal_behaviour.aepf <= aepf_at_ftp) & 
                                (pedal_behaviour.cpv > cpv_at_pref_cadence))
    time_in_quadrant_4  = (time_in_quadrant_4/float(pedal_behaviour.shape[0]))*100
    
    (fig, ax) = plt.subplots(figsize=(10,10), ncols=1, nrows=1)
    
    fig = sns.scatterplot(x="cpv", y="aepf", alpha=0.3, data=pedal_behaviour, ax=ax)
    
    sns.lineplot(x="cpv", y=power_curve_at_ftp, label="Power Curve at %.0f W"%ftp, 
                 color='orange', data=pedal_behaviour, ax=ax)
    ax.set(
        title='Force-Velocity Plot',
        xlabel='Circumferential Pedal Velocity (CPV) [m/s]',
           ylabel='Avg. Eff. Pedal Force (AEPF) [N]',
           xlim=(0, np.mean(pedal_behaviour.cpv)+np.std(pedal_behaviour.cpv)*3), 
           ylim=(0, np.mean(pedal_behaviour.aepf)+np.std(pedal_behaviour.aepf)*5))
    ax.axvline(cpv_at_pref_cadence, color='gray', linestyle="dashed")
    ax.axhline(aepf_at_ftp, color='gray', linestyle="dashed")
    
    ax.text(cpv_at_pref_cadence+0.1, aepf_at_ftp+10, 'Sprint: %.2f %%'%time_in_quadrant_1, fontsize=16, alpha=.5)
    ax.text(cpv_at_pref_cadence-0.8, aepf_at_ftp+10, 'Climb: %.2f %%'%time_in_quadrant_2, fontsize=16, alpha=.5)
    ax.text(cpv_at_pref_cadence-0.8, aepf_at_ftp-25, 'Recovery: %.2f %%'%time_in_quadrant_3, fontsize=16, alpha=.5)
    ax.text(cpv_at_pref_cadence+0.1, aepf_at_ftp-25, 'Crit: %.2f %%'%time_in_quadrant_4, fontsize=16, alpha=.5)
    
    return(fig)


plot_aepf_cvp_dist(workout.data, ftp, 80)
```
